```python
import re

def convert(value):
  if isinstance(value, str):
    match = re.search(r'([0-9.]+)k',value)
    if match:
      return  int(float(match.group(1))*1000)
    return value
  else:
    return value


for col in electricity.columns:
  electricity[col] = electricity[col].apply(convert)
```

```python
electricity['mean_usage'] = electricity.mean(axis=1)
oil['mean_usage'] = oil.mean(axis=1)
electricity = electricity.sort_values(by='mean_usage')
oil = oil.sort_values(by='mean_usage')
```

```python
electricity = electricity.drop(['mean_usage'],axis=1)
oil = oil.drop(['mean_usage'],axis=1)

plt.plot(oil.mean(axis=0))
plt.plot(oil.loc['Iran',:],'r--')
plt.legend(['World','Iran'])
plt.xlabel('Year')
plt.ylabel("Oil concumption per capita (Bl/year)")
plt.title("Oil consumption, Iran vs the World")
plt.grid()
```

```python
electricity = electricity.drop(['mean_usage'],axis=1)
oil = oil.drop(['mean_usage'],axis=1)

plt.plot(electricity.mean(axis=0))
plt.plot(electricity.loc['Iran',:],'r--')
plt.legend(['World','Iran'])
plt.xlabel('Year')
plt.ylabel("Electricity concumption per capita (KWh/year)")
plt.title("Electricity consumption, Iran vs the World")
plt.grid()
```

# **Phase 2**

```python
df = pd.read_csv('/content/Covid-19.csv')
display(df)
```

```python
select = df['countriesAndTerritories'] == 'Iran'
df_Iran = df.loc[select,:]
```

```python
x = df_Iran.dateRep.sort_index(ascending=False)
y_case = df_Iran.cases.sort_index(ascending=False)

plt.figure(figsize=(15,5))
plt.plot(x,y_case,'.')
plt.xticks(x[::31],rotation=45)
plt.xlabel("Date")
plt.ylabel("Cases")
plt.title("Cases trend during 2020")
plt.grid()
```

```python
df_Iran['expected_cases'] = (df_Iran['cases'].shift(-1)+df_Iran['cases'].shift(1))/2

x = df_Iran.dateRep.sort_index(ascending=False)
y_ex_case = df_Iran.expected_cases.sort_index(ascending=False)


plt.figure(figsize=(15,5))
plt.plot(x,y_ex_case,'.')
plt.xticks(x[::31],rotation=45)
plt.xlabel("Date")
plt.ylabel("Expexted Cases")
plt.title("Expected Cases trend during 2020")
plt.grid()
```

```python
x = df_Iran.dateRep.sort_index(ascending=False)
outlier = np.abs((df_Iran.expected_cases-df_Iran.cases)/df_Iran.expected_cases)


plt.figure(figsize=(15,5))
plt.plot(x,outlier,'.')
plt.xticks(x[::31],rotation=45)
plt.xlabel("Date")
plt.ylabel("ABS of Expexted Cases")
plt.title("ABS of Expected Cases trend during 2020")
plt.grid()
```

```python
x = df_Iran.dateRep.sort_index(ascending=False)
outlier = np.abs((df_Iran.expected_cases-df_Iran.cases)/df_Iran.expected_cases) > 0.7
y_case = df_Iran.cases.sort_index(ascending=False)
df_Iran.loc[outlier,'cases']=np.nan

plt.figure(figsize=(15,5))
plt.plot(x,y_case)
plt.xticks(x[::31],rotation=45)
plt.xlabel("Date")
plt.ylabel("ABS of Expexted Cases")
plt.title("ABS of Expected Cases trend during 2020")
plt.grid()
```

```python
x = df_Iran.dateRep.sort_index(ascending=False)
outlier = np.abs((df_Iran.expected_cases-df_Iran.cases)/df_Iran.expected_cases) > 0.7
y_case = df_Iran.cases.sort_index(ascending=False)
df_Iran.loc[outlier,'cases']=np.nan
y_case = y_case.interpolate(method='linear')

plt.figure(figsize=(15,5))
plt.plot(x,y_case)
plt.xticks(x[::31],rotation=45)
plt.xlabel("Date")
plt.ylabel("ABS of Expexted Cases")
plt.title("ABS of Expected Cases trend during 2020")
plt.grid()
```
